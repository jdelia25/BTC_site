```python
import pandas as pd
import altair as alt
```

```python
df_2 =  pd.read_excel("ethnicity_demo.xlsx")
summer_df_2 = df_2[df_2.term.str.contains('summer')]
summer_df_2 = summer_df_2.replace('American Indian/Alaskan', 'American Indian')

selection = alt.selection_multi(fields=['term'], bind='legend')
```

```python
colors = ['#1b9e77', '#d95f02', '#7570b3', '#e7298a', '#66a61e', '#e6ab02', '#a6761d', '#666666']
chart4 = alt.Chart(summer_df_2).mark_bar().encode(
    x=alt.X('count()', stack="normalize"),
    y='term',
    color='ethnicity'
).properties(
    title='Distribution of Ethnicity in Attending Students Over the Years'
).configure_range(
    category=alt.RangeScheme(colors)
)
```
